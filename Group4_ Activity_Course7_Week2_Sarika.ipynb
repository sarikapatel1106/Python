```python
import pandas as pd
import numpy as np
#df=pd.read_csv('Automobile_data.csv')
df=pd.read_csv('auto-mpg.csv')
df
```

```python
#1) Drop column Origin from the dataset.

df.drop(['origin'], axis=1,inplace=True)
df
```

```python
# 2) Find out all the missing values in the dataset and replace it with its most appropriate replacement.
df.isnull().sum()
df.replace("?", np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
# Display the name of the car with maximum number of horsepower.

#finding maximum horsepower first
max_horsepower=df['horsepower'].max()

df = df [['car name','horsepower']][df.horsepower==max_horsepower]
df
```

```python
# Q5) In our dataset, the fuel consumption columns "mpg" and are represented by mpg (miles per gallon) unit. 
# Assume we are developing an application in a country that accept the fuel consumption with L/100km standard. 
#change the name of column to "L/100km".
df.rename(columns={'mpg':'L/100km'}, inplace= True)
df
```

```python
# Q6) Normalize the weight column in this dataset.

df['weight'] = df['weight']/df['weight'].max()
df[['weight']].head(10)
```

```python
# Q6) In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it has a lot of unique values. 
#What if we only care about the price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? 
#Can we rearrange them into three â€˜bins' to simplify analysis?

#change data type of horsepower from object to int
df["horsepower"]=df["horsepower"].astype(int, copy= True)

bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins

group_names = ['Low Horsepower', 'Medium Horsepower', 'High Horsepower']

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)

df[['horsepower','horsepower-binned']].head(30)
```

```python
# Q7) Use a histogram to visualize the distribution of bins we created above.

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

#pyplot.bar(group_names, df["horsepower-binned"].value_counts())

plt.pyplot.hist(df["horsepower"], bins = 3)

# set labels  for axis and plot title
plt.pyplot.xlabel("Horsepower")
plt.pyplot.ylabel("Count")
plt.pyplot.title("Horsepower Bins")
```
