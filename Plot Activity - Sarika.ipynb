```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('company_sales_data.csv')
df.head()
```

```python
# 1. Exercise 1: Read Total profit of all months and show it using a line plot Total profit data provided for each month. 
# Generated line plot must include the following properties: â€“
# X label name = Month Number
# Y label name = Total profit

profit=df['total_profit'].tolist()
month=df['month_number'].tolist()

plt.plot(month, profit)
plt.rcParams["figure.figsize"] = (16,8)

plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.show()
```

```python
# 2. Exercise 2: Get total profit of all months and show line plot with the following Style properties
# Generated line plot must include following Style properties: â€“

# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3

profit=df['total_profit'].tolist()
month=df['month_number'].tolist()

plt.plot(month, profit, linestyle='--', color='r',
         label='Total Profit Last Year',
         markerfacecolor='r',
         marker='o',
         linewidth=3)

plt.rcParams["figure.figsize"] = (16,8)

plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')

plt.show()
```

```python
# Exercise 3: Read all product sales data and show it using a multiline plot
# Display the number of units sold per month for each product using multiline plots. 
# (i.e., Separate Plotline for each product).

month=df['month_number'].tolist()
facecream_data=df['facecream'].tolist()
facewash_data=df['facewash'].tolist()
toothpaste_data=df['toothpaste'].tolist()
bathingsoap_data=df['bathingsoap'].tolist()
shampoo_data=df['shampoo'].tolist()
moisturizer_data=df['moisturizer'].tolist()

plt.plot(month, facecream_data, label='Face Cream Sales Data for Last Year')
plt.plot(month, facewash_data, label='Face Wash Sales Data for Last Year')
plt.plot(month, toothpaste_data, label='Toothpaste Sales Data for Last Year')
plt.plot(month, bathingsoap_data, label='Bathing Soap Sales Data for Last Year')
plt.plot(month, shampoo_data, label='Shampoo Sales Data for Last Year')
plt.plot(month, moisturizer_data, label='Moisturizer Sales Data for Last Year')

plt.title('Sales Data by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')

plt.show()
```

```python
# 4. Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
# Also, add a grid in the plot. gridline style should â€œâ€“â€œ.

toothpaste_data=df['toothpaste'].tolist()
month=df['month_number'].tolist()

plt.scatter(month, toothpaste_data, label='Toothpaste Sales Data')
plt.grid(True, linewidth= 2, linestyle="--")

plt.title('Toothpaste Sales Data by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')

plt.show()
```

```python
# 5. Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

month=df['month_number'].tolist()
facecream_data=df['facecream'].tolist()
facewash_data=df['facewash'].tolist()

X = np.arange(len(month))

plt.bar(X - 0.2, facecream_data, 0.4, label='Face Cream Sales Data')
plt.bar(X + 0.2, facewash_data, 0.4, label='Face Wash Sales Data')

plt.title('Face Cream and Face Wash Sales Data by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')
plt.xticks(month)

plt.show()
```

```python
# 6. Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. 
# Save this plot to your hard disk

month=df['month_number'].tolist()
bathingsoap_data=df['bathingsoap'].tolist()

plt.bar(month, bathingsoap_data, label='Bathing Soap Sales Data')

plt.title('Bathing Sales Data by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')
plt.savefig('bathingsoap.png', dpi=120)

plt.show()
```

```python
# 7. Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

profit=df['total_profit'].tolist()
month=df['month_number'].tolist()

profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000, 400000]
plt.hist(profit, profit_range,label='Total Profit')

plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')

plt.show()
```

```python
# 8. Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

month=df['month_number'].tolist()
facecream_data=df['facecream'].tolist()
facewash_data=df['facewash'].tolist()
toothpaste_data=df['toothpaste'].tolist()
bathingsoap_data=df['bathingsoap'].tolist()
shampoo_data=df['shampoo'].tolist()
moisturizer_data=df['moisturizer'].tolist()

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
sales_data   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]

plt.pie(sales_data, labels=labels, autopct='%1.1f%%')

plt.title('Sales Data for Last Year')
plt.legend(loc='upper right')

plt.show()
```

```python
# 9. Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

month=df['month_number'].tolist()
facewash_data=df['facewash'].tolist()
bathingsoap_data=df['bathingsoap'].tolist()

f, axis_array = plt.subplots(2, sharex=True)
axis_array[0].plot(month, facewash_data, label='Face Wash Sales Data', color='r')
axis_array[0].set_title('Face Wash Sales Data')
axis_array[1].plot(month, bathingsoap_data, label='Bathing Soap Sales Data')
axis_array[1].set_title('Bathing Soap Sales Data')

plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')

plt.show()
```

```python
# 10.Exercise Question 10: Read all product sales data and show it using the stack plot

month=df['month_number'].tolist()
facecream_data=df['facecream'].tolist()
facewash_data=df['facewash'].tolist()
toothpaste_data=df['toothpaste'].tolist()
bathingsoap_data=df['bathingsoap'].tolist()
shampoo_data=df['shampoo'].tolist()
moisturizer_data=df['moisturizer'].tolist()

labels=['Face Cream','Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

fig, ax = plt.subplots()
ax.stackplot(month, facecream_data, facewash_data,toothpaste_data, bathingsoap_data, shampoo_data, moisturizer_data, 
             labels=labels)
plt.legend(loc='upper left')


plt.title('Sales Data by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')


plt.show()
```
